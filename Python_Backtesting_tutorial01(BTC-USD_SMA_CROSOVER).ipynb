```python
import yfinance as yf
import ta
import pandas as pd
from backtesting import Backtest,Strategy
from backtesting.lib import crossover
```

```python
class SMAcross(Strategy):
    st=50
    nd=100
    def init(self):
        close=self.data.Close
        self.sma1=self.I(ta.trend.sma_indicator,pd.Series(close),self.st)
        self.sma2=self.I(ta.trend.sma_indicator,pd.Series(close),self.nd)
    def next(self):
        if crossover(self.sma1,self.sma2):
            self.buy()
        elif crossover(self.sma2,self.sma1):
            self.sell()
        
        
```

```python
df=yf.download("BTC-USD",start="2020-01-01")
```

```python
df
```

```python
bt=Backtest(df,SMAcross,cash=100000,commission=0.002,exclusive_orders=True)
```

```python
output=bt.run()
output
```

```python
bt.plot()
```

```python
optim=bt.optimize(st=range(50,160,10),
                  nd=range(50,160,10),
                 constraint=lambda x: x.nd-x.st>20,
                  maximize="Return [%]")
bt.plot()
```
