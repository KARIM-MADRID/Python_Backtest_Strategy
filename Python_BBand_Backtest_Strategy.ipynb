```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=yf.download("MSFT",start="2018-01-01")
```

```python
df.head()
```

```python
df.index
```

```python
df["SMA_20"]=df.Close.rolling(window=20).mean()
df["SMA_50"]=df.Close.rolling(window=50).mean()
df
```

```python
df["stddev"]=df.Close.rolling(window=20).std()
df
```

```python
df["upper"]=df.SMA_20+2*df.stddev
df["lower"]=df.SMA_20-2*df.stddev
df["Buy_Signal"]=np.where(df.lower>df.Close,True,False)
df["Sell_Signal"]=np.where(df.upper<df.Close,True,False)
df
```

```python
df=df.dropna()
df
```

```python
#ploting 
plt.plot(df[["Close","SMA_20","SMA_50","upper","lower"]])
plt.show(figsize=(15,8))
```

```python
plt.plot(df[["Close","SMA_20","SMA_50","upper","lower"]])
plt.fill_between(df.index,df.upper,df.lower,color="grey",alpha=0.3,figsize=(15,8))
plt.show(figsize=(15,8))
```
